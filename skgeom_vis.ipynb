```python
from skgeom import *
```

```python
M = 20
T = 3
outer = [
    Segment2(Point2(-M, -M), Point2(-M, M)), Segment2(Point2(-M, M), Point2(M, M)),
    Segment2(Point2(M, M), Point2(M, -M)), Segment2(Point2(M, -M), Point2(-M, -M))
]

target = [
    Segment2(Point2(-T, -T), Point2(-T, T)), Segment2(Point2(-T, T), Point2(T, T)),
    Segment2(Point2(T, T), Point2(T, -T)), Segment2(Point2(T, -T), Point2(-T, -T))
]



# segments = [
#     Segment2(Point2(0, 0), Point2(0, 4)), Segment2(Point2(2, 4), Point2(8, 4)),
#     Segment2(Point2(3, 4), Point2(-8, 0)), Segment2(Point2(1, 0), Point2(0, 5)),
# ]
arr = arrangement.Arrangement()

for s in outer:
    arr.insert(s)

for s in target:
    arr.insert(s)
```

```python
for he in arr.halfedges:
    draw.draw(he.curve())
```

## Finding a point

```python
q = Point2(-2.5, 5)
face = arr.find(q)
print(face)
i = face.outer_ccb
first = next(i)
halfedge = next(i)

while first != halfedge:
    print(halfedge.source().point(), halfedge.target().point())
    halfedge = next(i)
print(halfedge.source().point(), halfedge.target().point())

```

## Computing visibility

```python
vs = RotationalSweepVisibility(arr)

# Halfedges are stored in doubly-connected lists and form chains. These chains define the inner and outer boundaries of connected components. 

q = Point2(0.4,0.4)
face = arr.find(q)
vis = vs.compute_visibility(q, face)

for he in arr.halfedges:
    draw.draw(he.curve(), visible_point = False)
for v in vis.halfedges:
    draw.draw(v.curve(), color='red', visible_point=False)
    print(v.source().point(), v.target().point())

draw.draw(q, color='magenta')
```
